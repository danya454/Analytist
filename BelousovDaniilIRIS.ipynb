```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn
url ='https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
d = np.genfromtxt(url, delimiter=',', dtype='str')
num=d[:,0:4]

#print(num)
nnum=num.astype(float)
print(nnum.min())
print(nnum.max())
print(nnum.mean())
print(np.median(nnum))
print(np.percentile(nnum,50)) #     
print(np.percentile(nnum,75))
print(np.percentile(nnum,25))

seaborn.set()

plt.style.use('fivethirtyeight')
plt.title('valuables for IRIS')
plt.xlabel('value of the 0 attribute')
plt.ylabel('count')
plt.tight_layout()
plt.hist(nnum[:,0],bins=50,edgecolor="black")
plt.show()

plt.style.use('fivethirtyeight')
plt.xlabel('value of the 1 attribute')
plt.ylabel('count')
plt.tight_layout()
plt.hist(nnum[:,1],bins=50,edgecolor="black")
plt.show()

plt.style.use('fivethirtyeight')
plt.xlabel('value of the 2 attribute')
plt.ylabel('count')
plt.tight_layout()
plt.hist(nnum[:,2],bins=50,edgecolor="black")
plt.show()

plt.style.use('fivethirtyeight')
plt.xlabel('value of the 3 attribute')
plt.ylabel('count')
plt.tight_layout()
plt.hist(nnum[:,3],bins=50,edgecolor="black")
plt.show()
```
